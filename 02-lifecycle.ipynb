## 2.2 Expected areas of highest effort
- Based on the dataset that I received, most effort is expected in the data cleaning and feature engineering.
The dataset has some missing values, especially in PV sensor readings (pv_mod1, pv_mod2 and pv_mod3), and the weather-related variables need to be checked and aligned to make sure everything matches correctly in time. This part will take more effort because the data must be reliable before any modelling can be done.
- Once the data is clean and well-structured, the modelling and optimization stages will be easier to perform.
- The lowest effort is expected in business framing and reporting, as these steps mainly involve interpreting the results and preparing the final outputs after analysis.

## 2.3 External data sources

The current dataset already includes energy demand, electricity price, PV generation, and several weather parameters such as temperature, pressure, cloud cover, wind speed, and radiation.
This information is sufficient for all required tasks - basic analysis, forecasting, and optimization tasks.

However, optional data sources could improve model performance:
- Calendar data (weekday, weekend, holidays) to capture behavioral patterns in demand.
- Updated weather forecasts for predictive modelling and short-term scheduling.

For this project, the existing dataset is absolutely adequate, but external weather and calendar data may be considered in later stages to enhance model generalization, but not necessary.
